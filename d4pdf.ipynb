```python
import pandas as pd
#import matplotlib.pyplot as plt
```

```python
#データがほしい座標を定義
x = 1
y = 0

#読み込むエクセルファイル（CSV形式）を指定
df = pd.read_csv('surf_HFB_4K_GF_m105_208608_smqr_subset.csv',header=None)
```

```python
#データを格納するためのからの配列
df4pdf_data = []

for t in range(0,744):
    data = df[x][y+155*t]
    df4pdf_data.append(data)
    print("t =" + str(t) + ":        "+str(data))

plt.figure(figsize=(20, 10))
plt.plot(df4pdf_data)

df4pdf_df = pd.Series(df4pdf_data,name='data')

#CSVに書き出し（ファイル名.CSV）
df4pdf_df.to_csv("data.csv")
```
